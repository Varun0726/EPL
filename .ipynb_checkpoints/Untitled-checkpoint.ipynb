```python
import requests 
from bs4 import BeautifulSoup
```

```python
Standings_url = "https://www.fotmob.com/leagues/47/overview/premier-league"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}

##https://www.statshub.com/league/premier-league/1
res = requests.get(Standings_url, headers=headers)
html = res.text
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
standing_table = soup.select('table.table')
```
